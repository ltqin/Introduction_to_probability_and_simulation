## What are sets? <a name="sets?"></a>

A set is a collection of distinct objects.

That is it don't think about it too much.

A very important set you will work with all the time is the **sample space** which is the set of all possible outcomes. We will denote the sample space with the Greek letter $\Omega$ (Omega).

## Set operations <a name="setoperations"></a>

We perform operations on numbers. For example, 69 + 89. We also perform opertations on sets.


Joining sets together is called the **Union**. We will denote a union the symbol $\cup$.

For example, let $A = \{3,7\}$ and $B = \{3,69\}$ then 
$$A \cup B = \{3,7,69\}.$$
Notice that we donot count '3' twice.

Exracting common elements is called the **intersection** We will denote a union the symbol $\cap$.
For example, 
$$A \cap B  = \{3\}$$
